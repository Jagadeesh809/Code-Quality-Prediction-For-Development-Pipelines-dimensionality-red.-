# CODE QUALITY PREDICTION IN DEVELOPMENT PIPELINES

dimensionality reduction

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python

# Sample dataset: Code metrics
data = {
    "Lines_of_Code": [100, 200, 150, 300, 400, 500, 600],
    "Cyclomatic_Complexity": [5, 10, 7, 20, 25, 30, 35],
    "Maintainability_Index": [75, 60, 70, 50, 40, 30, 20],
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python

# Preprocess data: Standardize features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python

# Dimensionality Reduction using PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions
pca_data = pca.fit_transform(scaled_data)
```

```python

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(pca_data)
```

```python

# Add cluster labels to the DataFrame
df['Quality_Cluster'] = clusters
```

```python

# Display results
print(df)

# Visualize clusters in the reduced dimensions
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df['Quality_Cluster'], cmap='viridis', s=100)
plt.title('Code Quality Clustering with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Quality Cluster')
plt.show()
```
